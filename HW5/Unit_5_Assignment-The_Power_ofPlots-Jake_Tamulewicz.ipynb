```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
from matplotlib import interactive
interactive(True)
import pandas as pd
import numpy as np

# File to Load (Remember to change these) - changed
city_data = "Resources/city_data.csv"
ride_data = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data)
ride_data_df = pd.read_csv(ride_data)

# Count number of rides per city for average fair calculation
# Then convert to dataframe and format
ride_count_sf = ride_data_df["city"].value_counts()
ride_count_df = pd.DataFrame(ride_count_sf)
ride_count_df = ride_count_df.reset_index()
ride_count_df.columns = ['city', 'ride_count']


# Get total of fares per city for average fair calculation
# Then convert to dataframe and format
total_fare_sf = ride_data_df.groupby('city')['fare'].sum()
total_fare_df = pd.DataFrame(total_fare_sf)
total_fare_df = total_fare_df.reset_index()
total_fare_df.columns = ['city', 'total_fare']

# Combine the data into a single dataset
combined_data_df = pd.merge(ride_count_df, total_fare_df, on=['city'])

# Calculate average fare per city and add column to combined data frame
avg_fare_sf = combined_data_df["total_fare"]/combined_data_df["ride_count"]
avg_fare_df = pd.DataFrame(avg_fare_sf)
avg_fare_df.columns = ['avg_fare']

# Add average fare per city to combined dataset
combined_data_df = combined_data_df.join(avg_fare_df, how='outer')

# Add city data to combined dataset
combined_data_df = pd.merge(combined_data_df, city_data_df, on=['city'])

# Display the data table for preview
combined_data_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

x = combined_data_df["ride_count"]
y = combined_data_df["avg_fare"]
z = combined_data_df["driver_count"]
z=z*z

# Build the scatter plots for each city types
plt.scatter(x, y, s=z, alpha=0.4, edgecolors="black", linewidth=2)
 
# Incorporate the other graph properties
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.title("Pyber Ride Sharing Data (2016)")

plt.show()
# Create a legend

# Incorporate a text label regarding circle size

# Save Figure
plt.show()
```
